# Exploratory Data Analysis

In this notebook, I'll perform a exploratory data analysis on the dataset before actually building the model


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
housing_prices = pd.read_csv("./Mumbai1.csv")
housing_prices.head()
```

```python
housing_prices.info()
```

As we can see from the previous code, there is no null data and the fields are mostly consisting of int64 data types.

```python
housing_prices.drop(columns=['Unnamed: 0'], axis=1, inplace=True)
```

```python
housing_prices.head()
```

```python
housing_prices.describe()
```

Most of the data is in a boolean format, yes or no, which can be translated as if the house has this feature or no. So, we'll tend to analyze the non-boolean characteristics, these include: Price, Area, No. of Bedrooms and Location (which is a categorical feature).

```python
mean_price_by_location = housing_prices.groupby(['Location']).agg({'Location' : ['count'], 'Price': ['mean']})
location = housing_prices[['Location']].value_counts()
```

```python

plt.scatter(x=location, y=mean_price_by_location)
```

```python
location_price_relationship = housing_prices.groupby(['Location']).agg({'Location' : ['count'], 'Price': ['mean']})
location_price_relationship.plot(kind="scatter", x=('Location', 'count'), y=('Price', 'mean'))
```

As we can see there isn't a clear relationship between the quantity of houses in a specific location and the mean price of them.

```python
quantity_of_houses_per_location = housing_prices.groupby(['Location']).agg({'Location' : ['count']})
quantity_of_houses_per_location.sort_values(by=('Location', 'count'), ascending=False)
```
